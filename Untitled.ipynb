```python
import pandas as pd
import numpy as np
data = pd.read_csv('worldcitiespop.csv', low_memory=False)
```

```python
data.head()
```

```python
CONVERSION_CONSTANT = 111

class CityDistace:
    
    def __init__(self, ref_point, columns):
        self.ref_longitude = ref_point[0]
        self.ref_latitude = ref_point[1]
        error_string = "logitude and or latitude columns missing in the dataset"
        try:
            self.latitude_column_name = [i for i in columns if i.lower()[: 3]=='lat'][0]
        except:
            raise Exception(error_string)

        try:
            self.longitude_column_name = [i for i in columns if i.lower()[: 3]=='lon'][0]
        except:
            raise Exception(error_string)

    def get_distance(self, df):
        latitude_values = df[self.latitude_column_name].values
        longitude_values = df[self.longitude_column_name].values
        distances = np.sqrt(((latitude_values - self.ref_latitude)*CONVERSION_CONSTANT)**2
                            +((longitude_values-self.ref_longitude)*CONVERSION_CONSTANT)**2)
        return distances
```

```python
kampala_cood = (0.3476, 32.5825)
col = ['Latitude', 'Longitude']
distance_from_kampala = CityDistace(ref_point=kampala_cood, columns=col)
```

```python
data['distance_from_kampala'] = distance_from_kampala.get_distance(data)
```
