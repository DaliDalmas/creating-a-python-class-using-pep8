```python
import pandas as pd
import numpy as np
data = pd.read_csv('worldcitiespop.csv',low_memory=False)
```

```python
CONVERSION_CONSTANT = 111

class CityDistance:
    def __init__(self, ref_point, columns):
        self.ref_logitude = ref_point[0]*CONVERSION_CONSTANT
        self.ref_latitude = ref_point[0]*CONVERSION_CONSTANT
        error_string = """Your columns should be a list conatining the dataframes longtude 
                  and latitude column names which should start with either lat or long"""
        try:
            self.latitude_column_name = [str(i) for i in columns if str(i).lower()[:3]=='lat'][0]
        except:
            raise Exception(error_string)    
        try:
            self.longitude_column_name = [str(i) for i in columns if str(i).lower()[:3]=='lon'][0]
        except:
            raise Exception(error_string)

    def get_distance(self, df):
        latitude_values = df[self.latitude_column_name].values
        longitude_values = df[self.longitude_column_name].values
        distances = np.sqrt((latitude_values*CONVERSION_CONSTANT-self.ref_logitude)**2\
                            +(longitude_values*CONVERSION_CONSTANT-self.ref_latitude)**2)
        return distances
```

```python
kampala_cood = (0.34, 32.58)
cols = ['Latitude', 'Longitude']
distance_from_kampala = CityDistance(ref_point=kampala_cood, columns=cols)
```

```python
data['distance_from_kampala'] = distance_from_kampala.get_distance(data)
```
